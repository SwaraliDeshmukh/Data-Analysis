```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_json('Dataframe.json')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df[((df['Quantity'] >= 6) & (df['Quantity'] <= 9))
      & (df['Rating'] < 3.0)
      & (df['Total Bill'] > df['Price'].quantile(0.95))]
```

```python
df.isnull().sum()
```

```python
def post_review(y):
    if y >= 4.0:
        return 'Good'
    elif y >= 3.5:
        return 'Average'
    else:
        return 'Bad'
df['Rating_category'] = df['Rating'].apply(post_review)
df[20:60]
```

```python
sns.lineplot(x="Type of item",y="Price",data=df)
df['Type of item'].describe()
```

```python
p=sns.relplot(x ="Rating", y ="Reviews",
            hue ="Type of item", data = df,aspect=2,height=6);

p.set(title = "Rating vs Reviews")
```

```python
fig , ax=plt.subplots(figsize=(20,5))
sns.set_context('notebook', font_scale = 2)
v=sns.barplot(x = 'Type of item', y = 'Rating', data = df,ax=ax)
v.set_title("Ratings for different categories of items")
plt.show()
```

```python

sns.catplot(x ='Quantity', y ='Price',
               hue = 'Type of item', data = df,kind="point",aspect=2,height=15)

plt.title("Price of items and their quantity")
plt.show()
```

```python
sns.pairplot(df,hue='Type of item')
```

```python
x=df.loc[df['Quantity'] > 5, 'Quantity']
y=df.loc[df['Quantity'] > 5, 'Reviews']
sns.regplot(x,y,data=df)
```

```python
y=df.loc[df['Price'] > 9000, 'Price']
x=df.loc[df['Price'] > 9000, 'Rating']
sns.regplot(x,y,data=df)
```

```python
graph = sns.FacetGrid(df, col ="Type of item")
graph.map(plt.scatter, "Total Bill", "Reviews").add_legend()

plt.show()
```

```python
fig , ax=plt.subplots(figsize=(30,10))
sns.histplot(data=df, x="Rating", hue="Type of item", multiple="stack",ax=ax,color='plasma')
```

```python
sns.catplot(
    data=df, x='Quantity', y='Total Bill', kind='box')
```
